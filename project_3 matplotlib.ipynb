```python
Q7=data.columns[data.columns.str.contains('Q7')]
```

```python
Q7
```

```python
darab[Q7]
```

```python
dic_Q7=dict()
for col in Q7:
    key=darab[col].value_counts().index[0]
    dic_Q7[key]=darab[col].value_counts()[0]    
```

```python
dic_Q7
```

```python
s=pd.Series(dic_Q7)
s
```

```python
plt.figure(figsize=(10,10))
plt.style.use('classic')
plt.pie(s, autopct="%1.1f%%",labels=s.index)
plt.title('a pie chart')
plt.legend(s.index)
plt.show()
```
